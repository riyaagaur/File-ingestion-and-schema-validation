#Step 2: Data Validation

Data validation ensures that the dataset conforms to expected standards and formats. This is a crucial step in the pipeline to identify and correct errors, inconsistencies, and anomalies in the data.


###Objectives of Data Validation:

Column Validation:

1.   Check if all required columns are present.
2.   Validate column names and their data types.


Content Validation:


1.   Identify and handle missing or null values.
2.   Remove invalid or inconsistent data (e.g., negative ages or salaries).

3.   Check for duplicate rows.

Range and Constraint Validation:


1.   Verify numeric values fall within expected ranges.
2.   Validate categorical values against predefined sets.

```python
required_columns = ['potential', 'value_eur', 'wage_eur', 'age', 'height_cm', 'weight_kg']
missing_columns = [col for col in required_columns if col not in chunk.columns]

if missing_columns:
    print(f"Missing columns: {missing_columns}")
else:
    print("All required columns are present.")
```

```python
chunk.columns = chunk.columns.str.replace(r'[^\w]', '_').str.strip()
print("Validated column names:", chunk.columns.tolist())
```

```python
print(chunk.isnull().sum())  # Count missing values per column

# Option 1: Fill missing values
chunk['value_eur'] = chunk['value_eur'].fillna(0)

# Option 2: Drop rows with missing values
chunk = chunk.dropna()
```

```python
expected_dtypes = {
    'potential': 'int64',
    'value_eur': 'float64',
    'wage_eur': 'float64',
    'age': 'int64',
    'height_cm': 'int64',
    'weight_kg': 'float64'
}

for col, dtype in expected_dtypes.items():
    if chunk[col].dtype != dtype:
        print(f"Converting {col} to {dtype}")
        chunk[col] = chunk[col].astype(dtype, errors='ignore')
```

```python
# Example: Validate ages
if (chunk['age'] < 16).any() or (chunk['age'] > 50).any():
    print("Invalid ages found.")
    chunk = chunk[(chunk['age'] >= 16) & (chunk['age'] <= 50)]

# Example: Validate wages
if (chunk['wage_eur'] < 0).any():
    print("Invalid wages found.")
    chunk = chunk[chunk['wage_eur'] >= 0]
```

```python
duplicate_rows = chunk.duplicated()
print(f"Number of duplicate rows: {duplicate_rows.sum()}")

chunk = chunk.drop_duplicates()
```

```python
!pip install pyyaml
```

```python
import yaml

# Load the schema from the YAML file
with open('/content/drive/MyDrive/Colab Notebooks/schema.yaml', 'r') as file:  # Provide the correct file path
    schema = yaml.safe_load(file)

print("Schema loaded successfully:")
print(schema)
```

```python
# Extract expected column names from the schema
expected_columns = [list(col.keys())[0] for col in schema['schema']['columns']]

# Check for missing columns
missing_columns = [col for col in expected_columns if col not in chunk.columns]

if missing_columns:
    print(f"Missing columns: {missing_columns}")
else:
    print("All required columns are present.")

```

```python
# Define a mapping between YAML data types and Python/pandas types
dtype_mapping = {
    'integer': 'int64',
    'float': 'float64',
    'string': 'object'
}

# Validate data types
for column in schema['schema']['columns']:
    col_name = list(column.keys())[0]  # Column name
    expected_type = column[col_name]['type']
    actual_type = str(chunk[col_name].dtype)

    # Check if the data type matches
    if actual_type != dtype_mapping[expected_type]:
        print(f"Column '{col_name}' has invalid type. Expected {dtype_mapping[expected_type]}, found {actual_type}.")
    else:
        print(f"Column '{col_name}' has a valid type.")
```

```python
# Example: Validate ranges
if 'age' in chunk.columns:
    if (chunk['age'] < 16).any() or (chunk['age'] > 50).any():
        print("Invalid age values found.")
        chunk = chunk[(chunk['age'] >= 16) & (chunk['age'] <= 50)]

# Example: Non-negative validation for wages
if 'wage_eur' in chunk.columns:
    if (chunk['wage_eur'] < 0).any():
        print("Invalid wage values found.")
        chunk = chunk[chunk['wage_eur'] >= 0]
```

```python
# Save the validated chunk
chunk.to_csv('/content/drive/MyDrive/validated_chunk.csv', mode='a', index=False)
print("Validated data saved successfully.")
```
